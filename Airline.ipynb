```python
#### one column we want to look into is flightNum, while it is numerical, it is technically categorical
### sometimes we may want to group data, for example Actual Elcapse Time, should we break them into Long Haul, Short Haul, etc?
### for TIme of departure/arrival, should we break into morning, noon, afternoon, etc? to have range
```

```python
### we create four buckets of time
ranges = [0, 600, 1200, 1800, 2400]
labels = ['Early Morning', 'Morning', 'Early Afternoon', 'Evening']
```

```python
### we can use cut function, similar to gradebook, 
##The category data type in pandas is a hybrid data type. 
##It looks and behaves like a string in many instances but internally is represented by an array of integers. 
##This allows the data to be sorted in a custom order and to more efficiently store the data.

##At the end of the day why do we care about using categorical values? There are 3 main reasons:

## We can define a custom sort order which can improve summarizing and reporting the data. In the example above, 
## “X-Small” < “Small” < “Medium” < “Large” < “X-Large”. Alphabetical sorting would not be able to reproduce that order.




df['DepTime2'] = pd.cut(df.DepTime, ranges, labels=labels).astype('category')
df['ArrTime2'] = pd.cut(df.ArrTime, ranges, labels=labels).astype('category')
```

```python
df.ArrTime2.head()
```

```python
### with category, we see top tells you what's the frequency
## describe returns a different set of data
df[['DepTime2', 'ArrTime2']].describe()
```

```python
### this provides same answer, we see morning is the most frequent one
df.DepTime2.value_counts()
```

```python
### now we are going to answer how often do planes get delayed based on their departure/arrival time
###We will group by 2 columns and only tract how much they delay

df.groupby(['DepTime2', 'ArrTime2'])[['DepDelay', 'ArrDelay']].mean()

### from data, we see for red eyes, leaving at even, arrive at morning, we see more delay on arrival
### from flights that leave morening and arrive at early morning, it is NaN, is it because that doesn't exist?
##What about early morning and arrive early afternoon, how is it possible that there is nothing?
```

```python
##how many items do we have that has early departure and early afternoon arrival
sum((df.DepTime < 600) & (df.ArrTime > 1200) & (df.ArrTime < 1600))
```

```python
### define both mean and standard deviation, as we want to know how big of a range does delay happens
agg_dict = {'ArrDelay': [np.mean, np.std], 'DepDelay': [np.mean, np.std]}


df.groupby(['DepTime2', 'ArrTime2']).agg(agg_dict)
### if we look at standard deviation, early morning to early morning is super unstable...very very large spread
### However, Early morning to morning is relatively stable

```

```python
#### then let's answer which flights are worst delay

df.head()

###let's create a new column called total delay time
#### we should look at difference for schedule time instead of actual
```

```python
### total delay time should be arr delay and dep delay add togehter
### convert schedule departure time into time buckets

df['TotalDelay'] = df.ArrDelay + df.DepDelay
df['ScheduledDepTime'] = pd.cut(df.CRSDepTime, ranges, labels = labels)
```

```python
####create a function to calculate average delay time for each dataframe, add that as a new column

def average_delay(dataframe):
    dataframe['AvgFlightDelay'] = dataframe.TotalDelay.mean()  ### create
    return dataframe
```

```python

avg_delayed = df.groupby('FlightNum').apply(average_delay)
#### we apply function to each flight number, each flight number as a unique group
```

```python
avg_delayed.head()
```

```python
### sort by values to see the worse delayed flright, we learned it is flight 2849
avg_delayed.sort_values('AvgFlightDelay', ascending=False)
```

```python
#### there is an easy way to do that through aggregation

df.groupby('FlightNum').agg({'TotalDelay': np.mean}).sort_values('TotalDelay', ascending=False).head()

### this shows the worst delay is flight 2849
```
