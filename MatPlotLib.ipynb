Sharex and sharey

```python
fig, ax = plt.subplots(2,1)
### We see this may be a bad idea, as x axis do not match up
ax[0].scatter(Wine.alcohol, Wine.heart)
ax[1].scatter(Wine.alcohol, Wine.liver)
```

```python
### add sharex is true to ensure they have same x axis
fig, ax = plt.subplots(2,1, sharex=True)

ax[0].scatter(Wine.alcohol, Wine.heart)
ax[1].scatter(Wine.alcohol, Wine.liver)
```

```python
### add sharex is true to ensure they have same x axis
fig, ax = plt.subplots(2,1, sharex=True)
##add some spacing between plots
fig.subplots_adjust(hspace = .4)

ax[0].scatter(Wine.alcohol, Wine.heart)
ax[1].scatter(Wine.alcohol, Wine.liver)
### Add chart titile
ax[0].set_title('Heart Disease')
ax[1].set_title('Liver Disease')

### Does it mean wine is good for your heart but bad for your livers?
```

# Wage Example

```python
W = pd.read_csv('wages.csv')
W.head()

##euro data statstics
```

```python
## Make year index
W = W.set_index('year')
W.head()
```

```python
###by default, it will use index as x axis
plt.plot(W.all_workers)
```

```python
### plot three lines in one plot
plt.plot(W.high_school)
plt.plot(W.college_degree)
plt.plot(W.advanced_degree)
```

```python
### Change color, line style to something more pleaseing
plt.plot(W.high_school, color = 'red', linestyle = '--')
plt.plot(W.college_degree, color = 'k', linestyle = ':')
## use tuples to decide spacing between dots
plt.plot(W.advanced_degree, color = '#aa55dd', linestyle = (0, (5,3,1,3)))
```

```python
### Use Markers
plt.plot(W.high_school, color = 'red', marker = 'o')
plt.plot(W.college_degree, color = 'k', marker = 'x')
## use tuples to decide spacing between dots
plt.plot(W.advanced_degree, color = '#aa55dd', marker = '*')
```

```python
### THere is shortcut for notations

### We want green, circle dot, dash...may not be intuitive
plt.plot(W.all_workers, 'go--')
```

```python
##by default, it is pretty easy to add legend, add lable and then use legend function
plt.plot(W.high_school, color = 'red', linestyle = '--', label = 'highschool')
plt.plot(W.college_degree, color = 'k', linestyle = ':', label = 'college degree')
plt.plot(W.advanced_degree, color = '#aa55dd', linestyle = (0, (5,3,1,3)), label = 'advance degree')
##legend
plt.legend()
```

Choose Colors

```python
#### picking color is not easy, what to do if you need to use a lot of colors

### matplotlib has a lot of color maps
##https://matplotlib.org/3.1.0/tutorials/colors/colormaps.html
```

```python
colormap = plt.cm.gnuplot
colormap(0.5)
```

```python
### create a quick expression of colors based on number of columns
colors = [colormap(i) for i in np.linspace(0,.9, W.shape[1])]
colors
```

```python
## drop all workers as it doesn't correlate with other columns

W2 = W.drop('all_workers', axis = 1)
```

```python
##show colors, we see a nature ordring of colors for legend

W2.plot(color = colors)

###While this is really good for quick data demo, there isn't too many flexibility
```

```python
#### Always good idea to create dictionary for labels, line style, color
#### For each column, I assigned specific label,line style and color

labeldict = {'less_than_high_school': 'less than high school', 
             'high_school': 'high school degree',
             'some_college': 'some college',
             'college_degree': 'college degree',
             'advanced_degree': 'advanced degree'
            }

linestyledict = {'less_than_high_school': (0,(1,1)), 
             'high_school': (0,(2,1)),
             'some_college': (0,(4,1)),
             'college_degree': (0,(6,1)),
             'advanced_degree': (0,(10,1))
            }

colordict = dict(zip(['less_than_high_school', 
             'high_school',
             'some_college',
             'college_degree',
             'advanced_degree'], colors))
```

```python
### use a for loop to plot all the variables
for var in W2.columns:
    plt.plot(W2[var], label = labeldict[var], color = colordict[var], linestyle = linestyledict[var])
plt.legend()
```

```python
## Moving Legend



fig, ax = plt.subplots(1,1)  ## use one charts
for var in W2.columns:
    ax.plot(W2[var], label = labeldict[var], color = colordict[var], linestyle = linestyledict[var])  ### Add plot to ax
ax.set_ylim((0,1600))

##location of legend
### bbox to anchor tells you where to put box, first number is in relation to x axis
#### original chart is from 0, to 1, so 1.01 means it would be outside of chart
#### second number is y locaiton, similarly, from 0 to 1
#### center left tells you which point from the legend box to align to the location you selected
ax.legend(bbox_to_anchor = (1.01, .5), loc = 'center left')


ax.set_xticklabels(range(2000,2019,2), rotation = 45)  ## show every 2 years, tilt lable by 45 degree
ax.set_ylabel('Weekly Wage')  ## add y label


### add the following
fmt = '${x:,.0f}'  ##show dollar size, with thousand comma
tick = mtick.StrMethodFormatter(fmt)
ax.yaxis.set_major_formatter(tick)
pass ### don't show a bunch of text
```
